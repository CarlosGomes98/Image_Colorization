```python
probs = np.load("prior_probs.npy")
np.sum(probs)
```

```python
uniform = np.zeros_like(probs)
uniform[probs!=0] = 1
uniform = uniform / np.sum(uniform)
np.sum(uniform)
```

```python
mix = (1-0.5) * probs + 0.5 * uniform
mix
```

```python
plt.plot(mix, label="mixed")
plt.plot(probs, label="original")
plt.legend()
plt.show()
```

```python
prior_factor = mix ** -1
prior_factor = prior_factor/np.sum(probs*prior_factor)
plt.plot(prior_factor)
plt.show()
```
