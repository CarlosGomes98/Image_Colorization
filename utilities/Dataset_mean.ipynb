```
from google.colab import drive
drive.mount('/content/drive')
import os
import numpy as np
from skimage import io, color
```

```
from keras.preprocessing.image import img_to_array, load_img, ImageDataGenerator
```

```
image_path = "/content/drive/My Drive/app/input/places"
image_size = 256
def read_image(img_id, dir):
    try:
        img = load_img(dir + "/" + img_id, target_size=(image_size, image_size))
        img = img_to_array(img)
        return img
    except:
        return None
```

```
image_count = 0
mean_pixels = np.zeros(5000)
for index, image in enumerate(os.listdir(image_path + "/Train")[0:5000]):
  print(index)
  img = read_image(image, image_path + "/Train")
  mean_pixels[image_count] = np.mean(color.rgb2lab(img*(1.0/255))[:, :, 0])
  image_count += 1
print(image_count)
mean_pixel = np.mean(mean_pixels)
print(mean_pixel)
```

```
f= open("mean.txt","w+")

f.write(mean_pixel)

f.close()
```
